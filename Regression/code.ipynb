```python
import pandas as pd

df = pd.read_csv('Boston House Price Data.csv')

# تمامی ستون‌های به عنوان ویژگی‌های ورودی (X)
features = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
X = df[features]

# این ستون برای رگرسیون خطی استفاده می‌شود.
y_price = df['PRICE']
```

رگرسیون خطی

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X_train, X_test, y_train_price, y_test_price = train_test_split(X, y_price, test_size=0.2, random_state=42)

model_linear = LinearRegression()
model_linear.fit(X_train, y_train_price)

print("--- نتایج رگرسیون خطی ---")
# این ضرایب نشان می‌دهند که هر ویژگی چقدر بر قیمت تأثیر می‌گذارد.
# مثلاً ضریب RM (تعداد اتاق) احتمالاً مثبت و بالا خواهد بود.
print(f"Coefficients (ضرایب): {model_linear.coef_}")
```

رگرسیون لجستیک

```python
df['High_Value'] = (df['PRICE'] > 22).astype(int)
y_value = df['High_Value']

X_train, X_test, y_train_value, y_test_value = train_test_split(X, y_value, test_size=0.2, random_state=42)

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model_logistic = LogisticRegression(solver='liblinear', random_state=0)
model_logistic.fit(X_train, y_train_value)

y_pred_value = model_logistic.predict(X_test)
accuracy = accuracy_score(y_test_value, y_pred_value)
#تابع accuracy فقط برای گسسته استفاده می شود

print("\n--- نتایج رگرسیون لجستیک ---")
print(f"Accuracy (دقت طبقه‌بندی گران/ارزان): {accuracy * 100:.2f}%")
```
